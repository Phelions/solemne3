### 1 - Naive Bayes

```python
from sklearn.model_selection import train_test_split
# Separo la columna y la rescato
X = np.array(df1.drop(['RainTomorrow'], 1))
y = np.array(df1['RainTomorrow'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 29)

# Define el algoritmo a utilizar Naive Bayes
from sklearn.naive_bayes import GaussianNB

modelo = GaussianNB()

# Entrenamiento del modelo
modelo.fit(X_train, y_train)

# ValidaciÃ³n del modelo
y_pred = modelo.predict(X_test)
```

```python
# EvaluaciÃ³n del modelo
from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(y_test, y_pred)
print('Matriz de ConfusiÃ³n:')
print(matriz)

# Se calcula la precisiÃ³n del modelo
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)
print('PrecisiÃ³n del modelo:', precision)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
print('PrecisiÃ³n  del modelo:',accuracy)
print('F1 score del modelo:',f1_score(y_test, y_pred))
print('Recall del modelo:',recall_score(y_test, y_pred))
```

### 3 - K-Means

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

X = np.array(df2[['Humidity9am','Humidity3pm','Cloud3pm']])
y = np.array(df2['RainTomorrow'])
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','pink']

asignar=[]

for row in y:
  asignar.append(colores[row])
ax.scatter(X[:, 2], X[:, 1], X[:, 0], c=asignar,s=3)
plt.show()


num_clusters = range(1, 10)
kmeans = [KMeans(n_clusters=i) for i in num_clusters]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(num_clusters,score, marker ='x')
plt.xlabel('NÃºmero de Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
lb = kmeans.predict(X)
C = kmeans.cluster_centers_
# 4 colores porque k = 4
colores=['red','green','blue','yellow']
asignar=[]
for row in lb:
  asignar.append(colores[row])
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=3)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=3)
plt.show()
                        
```
